```python
import json

%matplotlib inline
%config InlineBackend.figure_format='retina'
# import libraries
import numpy as np
import matplotlib as mp
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
#import slideUtilities as sl
#import laUtilities as ut
from importlib import reload
from datetime import datetime
from IPython.display import Image
from IPython.display import display_html
from IPython.display import display
from IPython.display import Math
from IPython.display import Latex
from IPython.display import HTML
print('')
```

```python
restaurant = {}
C = []

with open('dataset/business.json') as bus:
    for line in bus:
        dic = json.loads(line)
        if dic["city"]=="Las Vegas":
            if "Restaurants" in dic["categories"]:
                categories_list = [0,0,0,0,0,0,0,0]
                if "Mexican" in dic["categories"]:
                    categories_list[0]= int(1)

                if "Fast Food" in dic["categories"]:
                    categories_list[1]=1

                if "Pizza" in dic["categories"]:
                    categories_list[2]=1

                if "American (Traditional)" in dic["categories"]:
                    categories_list[3]=1
                 
                if "Sandwiches" in dic["categories"]:
                    categories_list[4]=1
                   
                if "Breakfast & Brunch" in dic["categories"]:
                    categories_list[5]=1
                  
                if "Chinese" in dic["categories"]:
                    categories_list[6]=1  
                    
                if "Italian" in dic["categories"]:
                    categories_list[7]=1  

                if categories_list != [0,0,0,0,0,0,0,0]:
                    C.append(categories_list)
                    restaurant[dic["business_id"]] = {"latitude":dic["latitude"], "longitude":dic["longitude"], "categories":dic["categories"],"popular":categories_list}

df = pd.DataFrame.from_dict(restaurant, orient='index')
df.reset_index(level=0, inplace=True)
df.head()
```

```python
latitude = df.latitude
longitude = df.longitude
location = np.array(list(zip(latitude, longitude)))


from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(1, 5),copy = True)
scaler.fit(location)
location = scaler.transform(location)
# print(location)
plt.scatter(location[:, 0],location[:, 1], s=10, alpha = 0.5)
plt.title('All restaurants distribution')
plt.xlabel('Scaled Latitude')
plt.ylabel('Scaled Longitude')
```

### Concatenate categories with laititude and longitude

```python
X = np.array(C)
X_train = np.concatenate((X, location), axis=1)
```

## K-Means Analysis

### Selecting the Number of Clusters

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import seaborn as sns; sns.set()  # for plot styling

max_clusters = 15
s = np.zeros(max_clusters+1)

for k in range(2,max_clusters+1):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(X_train)
    labels = kmeans.fit_predict(X_train)
    s[k] = metrics.silhouette_score(X_train,labels,metric='euclidean')

plt.plot(range(2,len(s)),s[2:])
plt.xlabel('Number of clusters')
plt.ylabel('Kmeans Silhouette Score')
plt.show()
```

### In the graph, we can see that when k = 11, kmeans Silhouette Score has the first peak. 
### So, I will select 11 as the number of clusters in the following Kmeans analysis

```python
k_K = 11
kmeans = KMeans(init='k-means++', n_clusters=k_K, n_init=10)
K_labels = kmeans.fit_predict(X_train)
K_centroids = kmeans.cluster_centers_

K_dic = {}
for i in K_labels:
    if i in K_dic:
        K_dic[i] += 1
    else:
        K_dic[i] = 1
        
print(K_dic)
for i in range(k_K):
    print('There are {} restautrants in cluster {}. '.format(K_dic[i], i))
    
```

```python
c = ['r', 'g', 'blue', 'yellow', 'purple','darkseagreen','dimgray' ,  'pink','brown','orange','beige','darkturquoise']

# plt.figure(figsize=(16,10))

for i in range(k_K):
    points = np.array([location[j] for j in range(len(location)) if K_labels[j] == i])
    plt.scatter(points[:, 0], points[:, 1], s=10, c=c[i], alpha = 0.5)

plt.title('Kmeans Clusters')
plt.xlabel('Scaled Latitude')
plt.ylabel('Scaled Longitude')
```

## Hierarchy Analysis

### Selecting the Number of Clusters

```python
import scipy.cluster
import scipy.cluster.hierarchy as hierarchy
import scipy.spatial.distance

H_Z = hierarchy.linkage(X_train, method='ward', metric='euclidean')

s = np.zeros(max_clusters+1)
for k in range(2,max_clusters+1):
    clusters = hierarchy.fcluster(H_Z, k, criterion='maxclust')
    s[k] = metrics.silhouette_score(X_train,clusters,metric='euclidean')
plt.plot(range(2,len(s)),s[2:])
plt.xlabel('Number of clusters')
plt.ylabel('Hierarchy Silhouette Score')
```

### In the graph, we can see that when k=8, kmeans Silhouette Score has the first peak. 
### So, I will select 8 as the number of clusters in the following Kmeans analysis

```python
k_H = 8
clusters = hierarchy.fcluster(H_Z, k_H, criterion='maxclust')

H_dic = {}
for i in clusters:
    if i in H_dic:
        H_dic[i] += 1
    else:
        H_dic[i] = 1
        
for i in range(1,k_H+1):
    print('There are {} restautrants in cluster {}. '.format(H_dic[i], i))
```

### Following figure is a hierarchy dendrogram.
### We can also see that there are 8 largest subtrees in the dendrogram, which means 8 clusters

```python
plt.figure(figsize=(14,4))
H_figure = hierarchy.dendrogram(H_Z,  truncate_mode='level', show_leaf_counts=True)
```

```python
c = ['r', 'g', 'blue', 'yellow', 'burlywood','darkseagreen', 'beige', 'pink', 'orange','dimgray','brown']

for i in range(k_H):
    points = np.array([location[j] for j in range(len(location)) if clusters[j] == i+1])
    plt.scatter(points[:, 0], points[:,1], c=c[i],s=10, alpha = 0.5)
plt.title('Hierachy Clusters')
plt.xlabel('Scaled Latitude')
plt.ylabel('Scaled Longitude')
```

## GMM Analysis

### Selecting the Number of Clusters

```python
from sklearn import mixture

s = np.zeros(max_clusters+1)
for k in range(2,max_clusters+1):
    gmm = mixture.GaussianMixture(n_components=k, covariance_type='full')
    gmm.fit(X_train)
    y_pred = gmm.predict(X_train)
    s[k] = metrics.silhouette_score(X_train,y_pred,metric='euclidean')
plt.plot(range(2,len(s)),s[2:])
plt.xlabel('Number of clusters')
plt.ylabel('GMM Silhouette Score')
```

### In the graph, we can see that when k=10, kmeans Silhouette Score has the first peak. 
### So, I will select 10 as the number of clusters in the following Kmeans analysis.

```python
k_G = 10
gmm = mixture.GaussianMixture(n_components=k_G, covariance_type='tied')
gmm.fit(X_train)
y_pred = gmm.predict(X_train)

G_dic = {}
for i in y_pred:
    if i in G_dic:
        G_dic[i] += 1
    else:
        G_dic[i] = 1
        
for i in range(k_G):
    print('There are {} restautrants in cluster {}. '.format(G_dic[i], i))
```

```python
c = ['r', 'g', 'blue', 'yellow', 'burlywood','darkseagreen', 'beige', 'pink', 'orange','dimgray','brown']
for i in range(k_G):
    points = np.array([location[j] for j in range(len(location)) if y_pred[j] == i])
    plt.scatter(points[:, 0], points[:, 1], s=10, c=c[i], alpha = 0.5)

plt.title('GMM Clusters')
plt.xlabel('Scaled Latitude')
plt.ylabel('Scaled Longitude')
```

d) Now let's detect outliers. These are the points that are the farthest from the centroids of their clusters. In order to find the outliers, you need to find the distances of all cluster points from the centroid and select the ones whose distance is above the 85th percentile. **(3 pts)**

## K-means clusters outliers

```python
from scipy.spatial import distance
from collections import defaultdict
import math

cluster_K = {}

for i in range(k_K):
    cluster_K[i] = np.array([X_train[j] for j in range(len(location)) if K_labels[j] == i])

dd = defaultdict(list)

for i in range(k_K):
    outlier = []
    for j in range(len(cluster_K[i])):
        d = distance.euclidean(cluster_K[i][j], K_centroids[i])
        dd[i].append(d)
    out_n = math.floor(0.15*len(dd[i]))
    print('There are {} outliers in K-Means cluster {} '.format(out_n, i))
    print('')
    d_sort = sorted(dd[i], reverse = True )
    for k in range(out_n):
        for j in range(len(cluster_K[i])):
            if dd[i][j] == d_sort[k]:
                outlier.append(cluster_K[i][j, 8:10])
    print("Their location (latitude, longitude) are {} ".format(outlier))
    print('')
    print('')
```

## Hierachy clusters outliers

```python
dd_H = defaultdict(list)

for i in range(k_H):
    outlier = []
    for j in range(len(cluster_H[i])):
        d = distance.euclidean(cluster_H[i][j], H_centroids[i])
        dd_H[i].append(d)
    out_n = math.floor(0.15*len(dd_H[i]))
    print('There are {} outliers in Hierachy cluster {} '.format(out_n, i+1))
    print('')
    d_sort = sorted(dd_H[i], reverse = True )
    for k in range(out_n):
        for j in range(len(cluster_H[i])):
            if dd_H[i][j] == d_sort[k]:
                outlier.append(cluster_H[i][j, 8:10])
    print("Their location (latitude, longitude) are {} ".format(outlier))
    print('')
    print('')
```

## GMM clusters outliers

```python
dd_G = defaultdict(list)

for i in range(k_G):
    outlier = []
    for j in range(len(cluster_G[i])):
        d = distance.euclidean(cluster_G[i][j], G_centroids[i])
        dd_G[i].append(d)
    out_n = math.floor(0.15*len(dd_G[i]))
    print('There are {} outliers in GMM cluster {} '.format(out_n, i))
    print('')
    d_sort = sorted(dd_G[i], reverse = True )
    for k in range(out_n):
        for j in range(len(cluster_G[i])):
            if dd_G[i][j] == d_sort[k]:
                outlier.append(cluster_G[i][j, 8:10])
    print("Their location (latitude, longitude) are {} ".format(outlier))
    print('')
    print('')
```

e) Give an analysis comparing the results you obtained from the 3 techniques. **(1 pts)**

1. K-means clusters change arbitrarily for different K, meanwhile Hierarchical clusters change is relatively stable.
2. K-means is sensative to outliers but relatively efficient :  time complexity O(nkdi),where n is the number of d-dimensional vectors, k the number of clusters and i the number of iterations needed until convergence. (reference:https://en.wikipedia.org/wiki/K-means_clustering)
3. Hierarchy is easy to get numbers of clusters. But it does not scale well: time complexity of at least O(n^2), where n is the number of total objects (reference: https://en.wikipedia.org/wiki/Hierarchical_clustering)
4. GMM works better with non-linear geometric distributions (best clusters distribution) than Kmeans because Kmeans calculate distance, while GM calculates weighted distance.(reference:https://en.wikipedia.org/wiki/Expectation%E2%80%93maximization_algorithm )
