```python
# The path should lead to your copy of chromedriver
driver = webdriver.Chrome('/Users/yt.hao/Desktop/BU/Study/CS506/cs506-spring2018-hw3-XintongHao-master/chromedriver')
url = "https://www.booking.com/searchresults.html?aid=304142&label=gen173nr-1FCAEoggJCAlhYSDNYBHIFdXNfbWGIAQGYATG4AQfIAQ3YAQHoAQH4AQKSAgF5qAID&sid=9f54d9b471e153a05becb7f53d06a3ed&class_interval=1&dest_id=20088325&dest_type=city&from_sf=1&group_adults=2&group_children=0&label_click=undef&lsf=class%7C5%7C72&nflt=di%3D929%3Bclass%3D5%3Breview_score%3D80%3B&no_rooms=1&pop_filter_id=review_score-80&pop_filter_pos=2&pop_filter_rank=3&raw_dest_type=city&room1=A%2CA&rows=15&sb_price_type=total&search_selected=1&src=index&ss=New%20York%2C%20New%20York%20State%2C%20USA&ss_raw=new%20york&ssb=empty"
# url = "https://www.booking.com/searchresults.html?label=gen173nr-1DCAEoggJCAlhYSDNYBHIFdXNfbWGIAQGYATG4AQfIAQ3YAQPoAQH4AQKSAgF5qAID&sid=991709bb93b40898e2ebcc9f2e20ade5&class_interval=1&dest_id=20088325&dest_type=city&from_sf=1&group_adults=2&group_children=0&label_click=undef&nflt=di%3D929%3B&no_rooms=1&pop_filter_id=di-929&pop_filter_pos=0&pop_filter_rank=2&raw_dest_type=city&room1=A%2CA&sb_price_type=total&search_selected=1&src=index&src_elem=sb&ss=New%20York%2C%20New%20York%20State%2C%20USA&ss_raw=new%20york&ssb=empty"
# Open the page in Selenium driver instance
driver.get(url)
```

```python
page_number = 1
while True:
    try:
        link = driver.find_element_by_link_text(str(page_number))
    except NoSuchElementException:
        break        
    driver.get(driver.current_url)
    # Find all hotels in the container
    for element in driver.find_elements_by_class_name("sr_item_no_dates"):
        # Get hotel name
        names = element.find_elements_by_class_name('sr-hotel__name')
        name = names[0]
        hotel_name.append(name.text)
        # Get rating
        # If no rating, return None
        if (len(element.find_elements_by_class_name('review-score-badge') )== 0):
            rating.append(None)
        else:
            rates = element.find_elements_by_class_name('review-score-badge')
            rate = rates[0]
            rating.append(rate.text)
        # Get class of rating
        # If no class of rating, return None
        if (len(element.find_elements_by_class_name('review-score-widget__text')) == 0):
            class_rating.append(None)
        else:
            class_rates = element.find_elements_by_class_name('review-score-widget__text')
            class_rate = class_rates[0]
            class_rating.append(class_rate.text)
    # While True, scrap next page
    page_number += 1
```

```python
df=pd.DataFrame(hotel_name,columns=['hotel_name'])
df['class_rating'] = class_rating
df['rating'] = rating
df.to_csv('hotel_information.csv', encoding='utf-8', index=False)
df.head()
```

```python
for rev_link in ex:
    page = urllib.request.urlopen(rev_link)
    soup = BeautifulSoup(page, 'lxml')
    hotel_name_box = soup.find('h2',  attrs={'class':'hp__hotel-name'})
    hotel_name = hotel_name_box.text.strip()
    
    reviews = soup.find_all('li', attrs={'class':'review_item clearfix '})

    for review in reviews:
        hotel_names.append(hotel_name)
        reviewer_box = review.find('div',attrs={'class':'review_item_reviewer'})
        reviewer_box = reviewer_box.find('h4')
        reviewer.append(reviewer_box.text.strip())
        print(reviewer_box.text.strip())

        if (review.find('div', attrs={'class': 'review_item_user_helpful_count'}) == None):
            num_vote.append(None)
            print(None)
        else:
            num_vote_box = review.find('div', attrs={'class': 'review_item_user_helpful_count'})
            num_vote.append(num_vote_box.text.strip()[:-13])
            print(num_vote_box.text.strip())

        date_box = review.find('p',attrs={'class':'review_item_date'})
        date.append(date_box.text.strip()[10:])
        print(date_box.text.strip()[10:])

        rating_box = review.find('span',attrs={'class':'review-score-badge'})
        rating.append(rating_box.text.strip())
        print(rating_box.text.strip())   

        if (review.find('p', attrs={'class': 'review_neg '}) == None):
            neg_review.append(None)
            print(None)
        else:
            neg_review_box = review.find('p', attrs={'class': 'review_neg '})
            neg_review.append(neg_review_box.text.strip()[1:])
            print(neg_review_box.text.strip()[1:])

        if (review.find('p', attrs={'class': 'review_pos '}) == None):
            pos_review.append(None)
            print(None)
        else:
            pos_review_box = review.find('p', attrs={'class': 'review_pos '})
            pos_review.append(pos_review_box.text.strip()[1:])
            print(pos_review_box.text.strip()[1:])

        print("============")
```
