```python
df=pd.DataFrame(hotel_names,columns=['hotel_name'])

df['reviewer_name'] = reviewer_names
df['date'] = dates
df['review_rating'] = review_ratings
df['neg_review'] = neg_reviews
df['pos_review'] = pos_reviews
df.to_csv('review_ratings.csv', encoding='utf-8', index=False)

df.head()
```
